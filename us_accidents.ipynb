```python
# object type to date type
file_raw_dates = file_raw.astype({'Start_Time':'datetime64[ns]'})
```

```python
# splitting datetime to date and time
file_raw_dates['start_date'] = file_raw_dates['Start_Time'].dt.date
file_raw_dates['start_time'] = file_raw_dates['Start_Time'].dt.time
file_raw_dates.dtypes
```

```python
# changing column order
file_dates = file_raw_dates[['ID', 'Source', 'TMC', 'Severity', 'Start_Time','start_date', 'start_time', 'End_Time',
       'Start_Lat', 'Start_Lng', 'End_Lat', 'End_Lng', 'Distance(mi)',
       'Description', 'Number', 'Street', 'Side', 'City', 'County', 'State',
       'Zipcode', 'Country', 'Timezone', 'Airport_Code', 'Weather_Timestamp',
       'Temperature(F)', 'Wind_Chill(F)', 'Humidity(%)', 'Pressure(in)',
       'Visibility(mi)', 'Wind_Direction', 'Wind_Speed(mph)',
       'Precipitation(in)', 'Weather_Condition', 'Amenity', 'Bump', 'Crossing',
       'Give_Way', 'Junction', 'No_Exit', 'Railway', 'Roundabout', 'Station',
       'Stop', 'Traffic_Calming', 'Traffic_Signal', 'Turning_Loop',
       'Sunrise_Sunset', 'Civil_Twilight', 'Nautical_Twilight',
       'Astronomical_Twilight']].astype({'start_date':'datetime64[ns]'})
```

```python
# extracting weekday from dates - example
def findDay(date): 
    born = datetime.datetime.strptime(date, '%Y-%m-%d').weekday() 
    return (calendar.day_name[born]) 
  
# Driver program 
date = '2020-12-24'
print(findDay(date))
```

```python
file_dates['start_date']
```

```python
file_dates['weekday'] = file_dates['start_date'].apply(lambda x: calendar.day_name[x.weekday()])
```

```python
file_dates
```

```python
order = ['Monday', 'Tuesday', 'Wednesday','Thursday','Friday','Saturday','Sunday']
file_dates.groupby(file_dates['weekday']).size().loc[order].plot.bar()
```
